 نمودار jointplot یکی از انواع نمودارهای ارتباطی در کتابخانه Seaborn می‌باشد. این نمودار برای نمایش توزیع دو متغیر به همراه رابطه‌ای بین آنها استفاده می‌شود. در این نمودار، دو نوع نمایش ارائه می‌شود:

نمودار پراکندگی (Scatter plot): در این بخش از نمودار، داده‌های دو متغیر به صورت نقاط بر روی یک صفحه نمایش داده می‌شوند و می‌تواند مناطقی با تراکم بالا یا پراکندگی خوبی از این داده‌ها نشان دهد.

نمودار توزیع هر متغیر (Marginal distributions): این بخش از نمودار، توزیع هر متغیر به صورت هیستوگرام یا توزیع چگالی احتمال را به طور جداگانه نمایش می‌دهد. این توزیع‌ها در لبه‌های و بالای یا سمت راست و چپ نمودار اصلی ظاهر می‌شوند.

موارد کاربرد اصلی jointplot عبارتند از:

بررسی ویژگی‌های دو متغیر و ارتباطی که بین آنها وجود دارد.
نشان دادن توزیع داده‌ها به صورت خلاصه و همچنین ارتباط بین متغیرها.
کشف الگوها، پدیده‌ها و تفاوت‌های مختلف در داده‌ها.
</p>

<font face="B Mitra">
<div dir=rtl>
<p align="justify">
<font size=5>
همانطور که مشاهده می‌شود، توزیع فواصل از محل اقامت تا محل کار و هزینه‌های حمل و نقل به نظر نزدیک به توزیع‌های نرمال می‌آید، در حالی که زمان غیبت به ساعت‌ها به شدت به سمت راست منحرف شده است. این موضوع باعث می‌شود مقایسه بین متغیرها دشوار باشد. یکی از راه‌حل‌های این مشکل، تبدیل داده‌ها به چیزی نزدیک به توزیع نرمال است. روشی موثر برای انجام این تبدیل، استفاده از تبدیل‌های Box-Cox یا Yeo-Johnson است که هر دو در کتابخانه scipy موجود است.

2.3. با توجه به اینکه ستون زمان غیبت شامل مقادیر صفر است از از تبدیل Yeo-Johnson استفاده کنید.
<br>
سپس هر دو نمودار بالا را مجددا رسم کنید و با تحلیل این این نمودارها درمورد تاثیرگذاری این دو فاکتور بر روی زمان غیبت توضیح دهید.
</p>
</font>
</div>
</font>

```python
from scipy.stats import yeojohnson

df["Absence_Duration_Hours"], lambda_value = yeojohnson(df["Absence_Duration_Hours"])

sns.jointplot(x='Absence_Duration_Hours', y='Distance_to_Work', data=df)
plt.show()
sns.jointplot(x='Absence_Duration_Hours',y='Transportation_Expense', data=df)
plt.show()
```

عملکرد اصلی نمودار Heatmap شامل موارد زیر است:

نمایش روابط میان داده‌ها: Heatmap به شما امکان می‌دهد تا روابط و الگوهای میان داده‌ها را به صورت بصری نشان دهید. با تحلیل رنگ‌های مختلف در heatmap، می‌توانید الگوهای مثبت یا منفی بین داده‌ها را تشخیص دهید.

تحلیل ماتریس‌ها: اگر داده‌های خود را به صورت یک ماتریس (جدول دو بعدی) دارید، از Heatmap می‌توان برای نمایش و تحلیل این ماتریس استفاده کرد. به سادگی می‌توانید تغییرات و الگوهای مختلف در داده‌ها را مشاهده کنید.

مقایسه اجزای مختلف: Heatmap به شما امکان می‌دهد تا اختلافات و فرق‌های مختلف بین اجزای مختلف داده‌ها را به صورت روشن و قابل تشخیص نشان دهید. این کمک می‌کند تا به سادگی بخش‌های مهم و تفاوت‌های قابل توجه را شناسایی کنید.
